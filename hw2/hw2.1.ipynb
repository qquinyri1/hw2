```python
%matplotlib inline

import numpy as np
import pandas as pd 
import matplotlib as plt

data_frame = pd.read_html(
                        "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8",
                       attrs={"class": "wikitable collapsible collapsed", "style": "text-align: right"},
                       match="Регіон"
                      )
table = data_frame[0]
table 
```

```python
table.head()
```

```python
table.shape
```

```python
table_upd_1 = table.replace(to_replace= "—", value= np.nan)
table_upd_1
```

```python
table_upd_1.dtypes
```

```python
table_upd_1["2014"] = pd.to_numeric(table_upd_1["2014"])
table_upd_1["2019"] = pd.to_numeric(table_upd_1["2019"])
table_upd_1.dtypes
```

```python
table_upd_1.isnull().sum()/table_upd_1.shape[0]
```

```python
table_upd_2 = table_upd_1.drop([27])
table_upd_2
```

```python
values = {"1950": table_upd_2["1950"].mean().round(1),
          "1960": table_upd_2["1960"].mean().round(1),
          "1970": table_upd_2["1970"].mean().round(1),
          "1990": table_upd_2["1990"].mean().round(1),
          "2000": table_upd_2["2000"].mean().round(1),
          "2012": table_upd_2["2012"].mean().round(1),
          "2014": table_upd_2["2014"].mean().round(1),
          "2019": table_upd_2["2019"].mean().round(1),
         }
table_upd_3 = table_upd_2.fillna(value=values)
table_upd_3
```

```python
avr_val2019 = table_upd_3["2019"].mean().round(1)
table_upd_3[table_upd_3["2019"] > avr_val2019]
```

```python
table_upd_4 = table_upd_3.sort_values(["2014"], ascending=False)[["Регіон", "2014"]].head(1)
table_upd_4
```

```python
graph = table_upd_3.plot.bar(
                            x="Регіон",
                            y="2019",
                            title = "Діаграма коефіцієнтів народжуваності в регіонах України у 2019 році",
                            ylabel = "Коефіцієнт",
                            figsize=(16, 10),
                            grid=True
)
graph
```

```python
diagram_1 = table_upd_3.plot(figsize=(14,5),
                        title='Коефіцієнт народжуваності в період з 1950-го по 2019-й рік',
                        xlabel='Індекс області',
                        ylabel='Коефіцієнт народжуваності')
```

```python
df_diagram_2 = table_upd_3.query("Регіон == 'Київ'")
df_diagram_2
```

```python
fig, ax = plt.subplots(figsize=(14, 5))
year = ['1950', '1960', '1970',	'1990',	'2000',	'2012',	'2014',	'2019']
koef = [231.0, 174.0, 159.0, 120, 73, 120, 121.0, 110.0]
plt.plot(year, koef, label='Коефіцієнт народжуваності', linestyle="-", marker="D", color="blue")
plt.xlabel('Рік', fontsize=15, color='black')
plt.ylabel('Коефіцієнт народжуваності', fontsize=15, color='black')
plt.title('Коефіцієнт народжуваності в м. Київ', fontsize=15)
plt.text(year[4], 75, 'Мінімальний коефіцієнт', color="red", fontsize=12)
plt.legend()
plt.show()
```
