```python
table_upd_1 = table.drop(labels=["Специализация", "Университет", "Предметная.область", "Валюта", "cls"], axis=1)
table_upd_1.isnull().sum()/table_upd_1.shape[0]
```

```python
table_upd_2 = table_upd_1.dropna()
table_upd_2.shape
```

```python
pydt = table_upd_2[table_upd_2['Язык.программирования'] == 'Python']
```

```python
pydt.shape
```

```python
pydt_upd_1 = pydt.groupby(by='Должность')
pydt_upd_1
```

```python
pydt_upd_2 = pydt_upd_1.agg({'Зарплата.в.месяц': ['min', 'max']})
pydt_upd_2
```

```python
def fill_avg_salary(data):
  return data.mean()

new_column_avg = pydt_upd_2.apply(fill_avg_salary, axis=1, result_type='expand')
new_column_avg.name = "avg_salary"
pydt_upd_3 = pd.concat([pydt_upd_2, new_column_avg], axis=1)
pydt_upd_3
```

```python
pydt_upd_3['avg_salary'].describe()
```

```python
pydt_upd_3.to_csv("2017_jun_final_updated.csv")
```
